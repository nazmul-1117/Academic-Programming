# Answer of Question 01

Implementation of DFS

```python
graph = {
    'Desktop': ['Semester_6'],
    'Semester_6': ['Desktop', 'AI'],
    'AI': ['Semester_6', 'Notes', 'Lab'],
    'Notes': ['AI'],
    'Lab': ['AI', 'DFS_Report.pdf'],
    'DFS_Report.pdf': ['Lab']
}

def search(graph, start, visited=None, depth=0, goal=None):
    if visited is None:
        visited = set()

    visited.add(start)
    print(f"{start}")

    if start == goal:
        print('-'*25)
        print(f"Goal '{goal}' found at Depth: {depth}")
        print('-'*25)
        return True

    for neighbor in graph[start]:
        if neighbor not in visited:
            if search(graph, neighbor, visited, depth + 1, goal):
                return True

    return False

search(graph, start = 'Desktop', goal = "DFS_Report.pdf")
```

# Answer of Question 02

- Implementation of CSP

```python
def is_valid(graph, coloring, vertex, color):
    for neighbor in graph[vertex]:
        if neighbor in coloring and coloring[neighbor] == color:
            return False
    return True

def graph_coloring_util(graph, coloring, vertices, m, idx):
    if idx == len(vertices):
        return True

    vertex = vertices[idx]
    for color in range(1, m+1):
        if is_valid(graph, coloring, vertex, color):
            coloring[vertex] = color
            if graph_coloring_util(graph, coloring, vertices, m, idx+1):
                return True
            coloring[vertex] = None

    return False

def graph_coloring(graph, m):
    vertices = list(graph.keys())
    coloring = {vertex: None for vertex in vertices}

    if graph_coloring_util(graph, coloring, vertices, m, 0):
        return coloring
    else:
        return None

graph = {
    'A' : ['B','D'],
    'B' : ['A', 'C'],
    'C' : ['B', 'F'],
    'D' : ['A', 'E'],
    'E' : ['D', 'F', 'G'],
    'F' : ['C', 'E'],
    'G' : ['E']
}

result = graph_coloring(graph, 3)
if result:
    print("Graph coloring solution:", result)
else:
    print("No solution found with 3 colors.")
```

# Answer of Question 03

```python
import math

def minimax(curDepth, nodeIndex, maxTurn, scores, targetDepth):
    if curDepth == targetDepth:
        return scores[nodeIndex]

    if maxTurn:
        return max(
            minimax(curDepth + 1, nodeIndex * 2, False, scores, targetDepth),
            minimax(curDepth + 1, nodeIndex * 2 + 1, False, scores, targetDepth)
        )
    else:
        return min(
            minimax(curDepth + 1, nodeIndex * 2, True, scores, targetDepth),
            minimax(curDepth + 1, nodeIndex * 2 + 1, True, scores, targetDepth)
        )

scores = []
n = int(input("Enter number of leaf nodes: "))
print("Enter the leaf node values:")
for _ in range(n):
    x = int(input())
    scores.append(x)


treeDepth = int(math.log2(len(scores)))

optimalValue = minimax(0, 0, True, scores, treeDepth)
print("The optimal value is:", optimalValue)
```
