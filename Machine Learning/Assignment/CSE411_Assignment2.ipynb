# Make Network Class

````python
class Network:
    def __init__(self, x: list[float], weights: list[float], target_output=0.5):
        self.x = x
        self.weights = weights
        self.target_output = target_output

        # nodes
        self.y3 = 0.0
        self.y4 = 0.0
        self.y5 = 0.0

        self.a3 = 0.0
        self.a4 = 0.0
        self.a5 = 0.0

        # deltas
        self.delta_3 = 0.0
        self.delta_4 = 0.0
        self.delta_5 = 0.0

    def forward_pass(self):
        # nodes
        self.a3 = self.x[0] * self.weights[0] + self.x[1] * self.weights[2]
        self.a4 = self.x[0] * self.weights[1] + self.x[1] * self.weights[3]

        # self.o1 = self.h1 * self.weights[4] + self.h2 * self.weights[5]
        self.y3 = self.sigmoid(self.a3)
        self.y4 = self.sigmoid(self.a4)

        self.a5 = self.y3 * self.weights[4] + self.y4 * self.weights[5]
        self.y5 = self.sigmoid(self.a5)

        return self.y5

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def error_calculate(self, y_hat=0.5):
        return self.target_output - y_hat

    def back_propagation(self):
        self.delta_5 = self.y5 * (1 - self.y5) * self.error_calculate(y_hat = self.y5)
        self.delta_4 = self.y4 * (1 - self.y4) * self.delta_5 * self.weights[5]
        self.delta_3 = self.y3 * (1 - self.y3) * self.delta_5 * self.weights[4]

        self.weights[4] += self.delta_5 * self.y3
        self.weights[5] += self.delta_5 * self.y4
        self.weights[3] += self.delta_4 * self.x[0]
        self.weights[2] += self.delta_4 * self.x[1]
        self.weights[1] += self.delta_3 * self.x[0]
        self.weights[0] += self.delta_3 * self.x[1]


    def print_weights(self):
        print(self.weights)
````

# My Assignment

````python
x = [0.43, 0.67]
weights = [0.17, 0.23, 0.3, 0.87, 0.53, 0.47]
target_op = 0.55

newtork = Network(x, weights, target_output=target_op)
# newtork.print_weights()

newtork.forward_pass()
newtork.back_propagation()
newtork.print_weights()
````
