```python
class GraphColoring:
    def __init__(self):
        self.V = 0
        self.num_colors = 0
        self.colors = []
        self.graph = []

    def graph_color(self, graph, num_colors):
        self.V = len(graph)
        self.graph = graph
        self.num_colors = num_colors
        self.colors = [0] * self.V

        if self.solve(0):
            print("\nSolution exists")
            self.display()
        else:
            print("No solution exists")

    def solve(self, vertex):
        if vertex == self.V:
            return True

        for color in range(1, self.num_colors + 1):
            if self.is_safe(vertex, color):
                self.colors[vertex] = color

                if self.solve(vertex + 1):
                    return True

                self.colors[vertex] = 0  # Backtrack

        return False

    def is_safe(self, vertex, color):
        for i in range(self.V):
            if self.graph[vertex][i] == 1 and self.colors[i] == color:
                return False
        return True

    def display(self):
        color_names = [
            "", "RED", "GREEN", "BLUE", "YELLOW",
            "ORANGE", "PINK", "BLACK", "WHITE"
        ]

        print("\nVertex Colors:")
        for i in range(self.V):
            print(f"Vertex {i} → {color_names[self.colors[i]]}")

def read_graph_from_file(filename):
    with open(filename, "r") as file:
        lines = file.readlines()

    V = int(lines[0].strip())
    graph = []

    for i in range(1, V + 1):
        row = list(map(int, lines[i].split()))
        graph.append(row)

    num_colors = int(lines[V + 1].strip())
    return graph, num_colors

if __name__ == "__main__":
    print("Graph Coloring lgorithm (File Input), Max-Color: 8\n")

    # filename = input("Enter input file name (e.g., graph.txt): ")
    filename = "graph.txt"

    graph, num_colors = read_graph_from_file(filename)

    gc = GraphColoring()
    gc.graph_color(graph, num_colors)
```
